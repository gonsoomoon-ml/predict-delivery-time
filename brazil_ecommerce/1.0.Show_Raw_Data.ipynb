```python
import pandas as pd
pd.options.display.max_rows = 5
from IPython.display import display

def show_df(file_name):
    df = pd.read_csv(file_name)
    display(df)
    return df
    
```

## Customer

```python
file_name = 'data/olist_customers_dataset.csv'    
df = show_df(file_name)     
```

```python
df.nunique()
```
